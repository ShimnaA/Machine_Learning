*** One hot encoding with Many Categories ***

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('train.csv')
```

```python
data.head()
```

```python
data = pd.read_csv('train.csv', usecols= ['X1', 'X2','X3', 'X4','X5', 'X6'] )
```

```python
data.head()
```

```python
data.shape
```

```python
for col in data.columns:
    print(col + "  :  " + str(data[col].nunique()) + " labels")
```

```python
pd.get_dummies(data, drop_first=True)
```

```python
data['X2'].value_counts().sort_values(ascending=False).head(20)
```

```python
# List of all the most frequent 10 categories of X2 column
top_10 = [x for x in data['X2'].value_counts().sort_values(ascending=False).head(10).index]
```

```python
top_10
```

```python
#Making 10  binary variables

for label in top_10:
    data[label] = np.where(data['X2']==label, 1, 0)
```

```python
data
```

```python
def one_hot_top_x(df, variable, top_x_labels):
    # function to create the dummy variables for the most frequent tables
    for label in top_x_labels:
        df[variable + '_' + label] = np.where(df[variable]==label, 1, 0)
    

one_hot_top_x(data, 'X2', top_10)
```

```python
data
```

```python
data = pd.read_csv('train.csv', usecols= ['X1', 'X2','X3', 'X4','X5', 'X6'] )
```

```python
one_hot_top_x(data, 'X2', top_10)
```

***Advantages***
    - Straight forward
    - does not require hrs of variable exploration
    - does not expand massively the feature space

***Disadvantages***
    - Does not add any info that may make the variable more predictive
    - does not keep the information of the ignored labels
