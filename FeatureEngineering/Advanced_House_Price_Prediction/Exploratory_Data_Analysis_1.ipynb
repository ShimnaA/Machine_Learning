#### Missing Values

```python
## Here we will check the percentage of nan values present in each feature

train.isnull().sum()
```

```python
train.isnull().mean()
```

```python
features_with_na = [feat for feat in train.columns if train[feat].isnull().sum()>0 ]
```

```python
features_with_na
```

```python
for feat in features_with_na:
    print("null values% of  {} : {}  ".format(feat, round(train[feat].isnull().mean(),4)))
```

### Since they are many missing values, we need to find the relationship between missing values and Sales Price

Let's plot some diagram for this relationship


```python
for feature in features_with_na:
    data = train.copy()
    data[feature] = np.where(data[feature].isnull(), 1, 0)
    
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.title(feature)
    plt.show()
```

###### The relation between the missing values and the output variable is clearly visible, hence we need to replace the missing values with something meaningful in feature engineering phase.

### Numerical Features

```python
numerical_features = [feature for feature in train.columns if train[feature].dtypes != 'O']
print("Number of Numerical Featues is {}".format(len(numerical_features)))
train[numerical_features].head(2)
```

#### Temporal Variables(Eg: Datetime Variables)

From the Dataset we have 4 year variables. We have extract information from the datetime variables like no of years or no of days. One example in this specific scenario can be difference in years between the year the house was built and the year the house was sold.

```python
train.columns
```

```python
year_features = [feature for feature in train.columns if 'Yr' in feature or 'Year' in feature ]
year_features
```

```python
train[year_features]
```

```python
for feature in year_features:
    print(feature, train[feature].unique())
```

### Analyse the temporal vairables, check whether any relation between year sold and Price

```python
train.groupby('YrSold')['SalePrice'].median().plot()
plt.xlabel("Year Sold")
plt.ylabel("Median House Price")
plt.title("House Price vs Year Sold")
```

### we compare all the year features and sale price

```python
train.groupby('YearRemodAdd')['SalePrice'].median().plot()
plt.xlabel("Year Remodified")
plt.ylabel("Median House Price")
plt.title("House Price vs Year Remodified")
```

```python
train.groupby('YearBuilt')['SalePrice'].median().plot()
plt.xlabel("Year Built")
plt.ylabel("Median House Price")
plt.title("House Price vs Year Built")
 
```

```python
### Compare the difference between the YEar sold and the other year features
data = train.copy()
for feature in year_features:
    
    if feature == 'YrSold': continue
    data[feature] = data['YrSold'] - data[feature]
    plt.scatter(data[feature], data['SalePrice'])
    plt.title(feature)
    plt.show()
```

```python
## Numerical variables are usually of 2 type
## 1. Continous variable and Discrete Variables

numerical_features
discrete_numerical_features = [feature for feature in numerical_features if train[feature].nunique()<25 and feature not in year_features + ['Id']]

print("Number of Numerical Features are {}".format(len(discrete_numerical_features)))
```

```python
discrete_numerical_features
```

```python
train[discrete_numerical_features].head(3)
```

```python
## Lets Find the realtionship between them and Sale PRice


for feature in discrete_numerical_features:
    data = train.copy()
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.title(feature)
    plt.show()
```

#### There is a relationship between variable number and SalePrice

## Continuous Features

```python
continuous_features = [feature for feature in numerical_features if feature not in discrete_numerical_features + year_features + ['Id'] ]
print("Number of continuous features are {}".format(len(continuous_features)))
```

```python
##Lets understand the continuous feature by plotting histogram
data = train.copy()
for feature in continuous_features:
    data[feature].hist(bins=30)
    plt.title(feature)
    plt.show()
```
