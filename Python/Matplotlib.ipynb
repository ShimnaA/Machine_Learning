```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.arange(0,10)
```

```python
y = np.arange(11,21)
```

```python
a = np.arange(40,50)
b = np.arange(50,60)
```

```python
## scatter plot
plt.scatter(x,y,color='b')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Graph in 2D')
plt.savefig('Test.png')
```

```python
#plot
y = x**2
plt.plot(x,y, 'ro--')
```

```python
plt.plot(x,y,'r*',linestyle='dashed',linewidth=2, markersize=12)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('2d Diagram')
```

```python
#subplots  - Create mulitple plots in one canvas
plt.subplot(1,3,1)
plt.plot(x,y,'g*')
plt.subplot(1,3,2)
plt.plot(y,x,'ro')
plt.subplot(1,3,3)
plt.plot(x,x,'b--')
```

```python
print(np.pi)
x = np.arange(0,5 * np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)
```

```python
plt.plot(x,y_sin, 'g')
```

```python
plt.subplot(2,1,1)
plt.plot(x, y_sin, 'g*')
plt.subplot(2,1,2)
plt.plot(x, y_cos, 'r*')
```

```python
#BarPlot
x = [1,2,3]
y = [5,3,5]
plt.bar(x,y)
x2 = [6,7,8]
y2 = [5,3,5]
plt.bar(x2,y2)
```

```python
#histograms
a =  np.array([1,2,3,3,4,3,3,4,2,2,1,2,3,2,3,2,1])
plt.hist(a)
```
